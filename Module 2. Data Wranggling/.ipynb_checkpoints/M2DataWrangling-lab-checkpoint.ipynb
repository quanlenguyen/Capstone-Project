## Imputing missing values


Find the  value counts for the column WorkLoc.


```python
# your code goes here
workloc_count=df['WorkLoc'].value_counts()
workloc_count
```

Identify the value that is most frequent (majority) in the WorkLoc column.


```python
#make a note of the majority value here, for future reference
majority_value=workloc_count.idxmax()
```

Impute (replace) all the empty rows in the column WorkLoc with the value that you have identified as majority.


```python
# your code goes here
df['WorkLoc'].replace(np.nan, majority_value,inplace=True)
```

After imputation there should ideally not be any empty rows in the WorkLoc column.


Verify if imputing was successful.


```python
# your code goes here
df[df['WorkLoc'].isnull()==True].shape[0]
```

## Normalizing data


There are two columns in the dataset that talk about compensation.

One is "CompFreq". This column shows how often a developer is paid (Yearly, Monthly, Weekly).

The other is "CompTotal". This column talks about how much the developer is paid per Year, Month, or Week depending upon his/her "CompFreq". 

This makes it difficult to compare the total compensation of the developers.

In this section you will create a new column called 'NormalizedAnnualCompensation' which contains the 'Annual Compensation' irrespective of the 'CompFreq'.

Once this column is ready, it makes comparison of salaries easy.


<hr>


List out the various categories in the column 'CompFreq'


```python
# your code goes here
marjority_value=df['CompFreq'].value_counts().idxmax()
df['CompFreq'].replace(np.nan,marjority_value, inplace=True)
df['CompTotal'].replace(np.nan,df['CompTotal'].mean(), inplace=True)
df['CompFreq'].unique()
```

Create a new column named 'NormalizedAnnualCompensation'. Use the hint given below if needed.


```python
def normalize_compensation(a):
    if a['CompFreq'] == 'Yearly':
        return a['CompTotal']
    elif a['CompFreq'] == 'Monthly':
        return a['CompTotal'] * 12
    elif a['CompFreq'] == 'Weekly':
        return a['CompTotal'] * 52
    else:
        return None

df['NormalizedAnnualCompensation'] = df.apply(normalize_compensation, axis=1)
```

```python
df[['CompFreq','CompTotal','NormalizedAnnualCompensation']]
```

## Demo: How to work with database


Download database file.


```python
file= "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m4_survey_data.sqlite"
```

```python
import urllib
urllib.request.urlretrieve(file, "m4_survey_data.sqlite")
```

Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


```python
df=pd.read_sql_query("""SELECT * FROM master""", conn)
df
```

Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import numpy as np
import matplotlib.pyplot as plt
df_histogram=df['ConvertedComp'].dropna()
bins,bin_edges=np.histogram(df_histogram,15)
df_histogram.plot(kind='hist', bins=15)
plt.title('Histogram Graph of ConvertedComp')
plt.xticks(bin_edges, rotation=40)
plt.show()
bin_edges
```

### Box Plots


Plot a box plot of `Age.`


```python
def upper_lower(data): #data is a series of dataframe
    Q3=data.quantile(q=0.75)
    Q1=data.quantile(q=0.25)
    IQR=Q3-Q1
    Upper=Q3+1.5*IQR
    Lower=Q1-1.5*IQR
    return Upper, Lower
```

```python
# your code goes here
df_box=df['Age'].dropna()
upper_lower_age=upper_lower(df_box)
df_box_clean=df_box[(df_box>upper_lower_age[1]) & (df_box<upper_lower_age[0])]
df_box_clean.plot(kind='box')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
import seaborn as sns
sns.scatterplot(data=df,x='Age',y='WorkWeekHrs')
```

```python
# your code goes here
import seaborn as sns
upper_lower_workHrs=upper_lower(df['WorkWeekHrs'])
df_scatter=df[['Age','WorkWeekHrs']]
df_scatter_clean=df_scatter[(df_scatter['WorkWeekHrs']<upper_lower_workHrs[0])&(df_scatter['WorkWeekHrs']>upper_lower_workHrs[1])&(df_scatter['Age']<100)]
sns.scatterplot(data=df_scatter_clean,x='Age',y='WorkWeekHrs')
plt.title('Scatter plot of Age and WorkWeekHrs')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
sns.scatterplot(data=df,x='WorkWeekHrs',y='CodeRevHrs')
```

```python
# your code goes here
df_bubble=df[['WorkWeekHrs','CodeRevHrs','Age']]
df_bubble_clean=df_bubble[(df_bubble['WorkWeekHrs']<100)&(df_bubble['CodeRevHrs']<20)]
sns.scatterplot(data=df_bubble_clean,x='WorkWeekHrs',y='CodeRevHrs',size='Age')
plt.title('Bubble plot of CodeRevHrs and WorkWeekHrs')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
df2=pd.read_sql_query("""SELECT * FROM DatabaseDesireNextYear""", conn)
df2
```

```python
df_pie=df2['DatabaseDesireNextYear'].value_counts()
plt.figure(figsize=(16,10))
plt.pie(df_pie, labels=df_pie.index, autopct='%1.1f%%',shadow=True, pctdistance=.8)
plt.title('Top 5 Databases most wish to learn next year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
df_stacked=df.groupby('Age')[['WorkWeekHrs','CodeRevHrs']].median().loc[30:35,:]
df_stacked.T.plot(kind='bar',stacked='True')
plt.legend
df_stacked
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
df_line=df.groupby('Age')['ConvertedComp'].median().to_frame().loc[45:60,:]
sns.lineplot(df_line,x=df_line.index,y=df_line['ConvertedComp'])
plt.title('Median ConvertedComp for all ages from 45 to 60.')
plt.show()
```
